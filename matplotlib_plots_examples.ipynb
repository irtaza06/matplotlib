```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
sns.set_style("ticks")
sns.set_context("poster")
%pylab inline
```

# Baisc plot

```python
figsize = (6,5)
tickslabel_size = 18
fig = plt.figure(figsize=figsize)
plt.rc('font', family='arial')
plt.rc('xtick', labelsize=tickslabel_size)
plt.rc('ytick', labelsize=tickslabel_size)
plt.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
plt.xlabel('Angle/rad')
plt.ylabel('sin(x)')
```

# Multirow plot

```python
#multi row plot
figsize = (12,6)
tickslabel_size = 18
plt.rc('font', family='arial')
plt.rc('xtick', labelsize=tickslabel_size)
plt.rc('ytick', labelsize=tickslabel_size)
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=figsize,sharex=True,sharey=True, constrained_layout=True)
for ax in axes:
    ax.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
    ax.set_ylabel('sin(x)')
ax.set_xlabel('Angle/rad')
```

# Multirow, multicolumn plot

```python
#multi row, multi column plot
figsize = (12,6)
tickslabel_size = 18
plt.rc('font', family='arial')
plt.rc('xtick', labelsize=tickslabel_size)
plt.rc('ytick', labelsize=tickslabel_size)
fig,axes = plt.subplots(nrows=3,ncols=2, figsize=figsize,sharex=True,sharey=True, constrained_layout=True)
#looping through rows
for r in range(axes.shape[0]):
    #looping through rows
    for c in range(axes.shape[1]):
        #print(r,c)
        axes[r,c].plot(np.sin(np.linspace(-np.pi, np.pi,100)))
        axes[r,c].text(50,1.2,str(r)+str(c),fontsize=tickslabel_size)
        if (c==0):
            axes[r,c].set_ylabel('sin(x)')
        if (r==axes.shape[1]):
            axes[r,c].set_xlabel('Angle/rad')
```

# Custom plot | Easy way

```python
#choose between "notebook", "article", "poster" and "talk" 
sns.set_context("poster")

figsize = (12,16)
tickslabel_size = 18
plt.rc('font', family='arial')
plt.rc('xtick', labelsize=tickslabel_size)
plt.rc('ytick', labelsize=tickslabel_size)
fig = plt.figure(figsize=figsize)
#Define grid shape, think of it as drawing horizontal and vertical lines like traditional graph papers
#First argument of shape is horizontal lines
#Second argument of shape is vertical  lines
shape = (6, 5)
#Second argument (0,0) is location
ax1 = plt.subplot2grid(shape, (0, 0), rowspan=1,colspan=5)
ax1.text(50,1.2,'ax1',fontsize=tickslabel_size)
ax1.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax1.set_xlabel('Angle/rad')
ax1.set_ylabel('sin(x)')

ax2 = plt.subplot2grid(shape, (1, 2), rowspan=1,colspan=3)
ax2.text(50,1.2,'ax2',fontsize=tickslabel_size)
ax2.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax2.set_xlabel('Angle/rad')
ax2.set_ylabel('sin(x)')

ax3 = plt.subplot2grid(shape, (1, 0), rowspan=5,colspan=2)
ax3.text(50,1.2,'ax3',fontsize=tickslabel_size)
ax3.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax3.set_xlabel('Angle/rad')
ax3.set_ylabel('sin(x)')


ax4 = plt.subplot2grid(shape, (2, 2), rowspan=1,colspan=3,sharex=ax2)
ax4.text(50,1.2,'ax4',fontsize=tickslabel_size)
ax4.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax4.set_xlabel('Angle/rad')
ax4.set_ylabel('sin(x)')


ax5 = plt.subplot2grid(shape, (3, 2), rowspan=3,colspan=3,sharex=ax2)
ax5.text(50,1.2,'ax5',fontsize=tickslabel_size)
ax5.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax5.set_xlabel('Angle/rad')
ax5.set_ylabel('sin(x)')
fig.tight_layout()


 

# plt.plot(np.linspace(-1,1,100))
# plt.subplot2grid(shape, (1, 0), rowspan=1,colspan=1)
# plt.plot(np.linspace(-1,1,100))
```

# Custom plot | Efficient way

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})

figsize = (12,12)
tickslabel_size = 18
plt.rc('font', family='arial')
plt.rc('xtick', labelsize=tickslabel_size)
plt.rc('ytick', labelsize=tickslabel_size)
fig = plt.figure(figsize=figsize,constrained_layout=True)
#Similarly define grid by pass # of rows and # of cols
gs = fig.add_gridspec(6,5)
#Now just like list slicing, it is very flexible to make desired layout bu slicing gs(grdispec)
ax1 =  fig.add_subplot(gs[0, :])
ax1.text(50,1.2,'ax1',fontsize=tickslabel_size)
ax1.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax1.set_xlabel('Angle/rad')
ax1.set_ylabel('sin(x)')

ax2 =  fig.add_subplot(gs[1:2, 3:])
ax2.text(50,1.2,'ax2',fontsize=tickslabel_size)
ax2.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax2.set_xlabel('Angle/rad')
ax2.set_ylabel('sin(x)')

ax3 =  fig.add_subplot(gs[1:, 0:3])
ax3.text(50,1.2,'ax3',fontsize=tickslabel_size)
ax3.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax3.set_xlabel('Angle/rad')
ax3.set_ylabel('sin(x)')

ax4 =  fig.add_subplot(gs[2:3, 3:])
ax4.text(50,1.2,'ax4',fontsize=tickslabel_size)
ax4.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax4.set_xlabel('Angle/rad')
ax4.set_ylabel('sin(x)')

ax5 =  fig.add_subplot(gs[3:, 3:])
ax5.text(50,1.2,'ax5',fontsize=tickslabel_size)
ax5.plot(np.sin(np.linspace(-np.pi, np.pi,100)))
ax5.set_xlabel('Angle/rad')
ax5.set_ylabel('sin(x)')
fig.tight_layout()
#hide right y and upper x
sns.despine()
# To save and remvoe extra while space around the plot use: bbox_inches='tight'
plt.savefig('figure.png',dpi=300,bbox_inches='tight')
```
